Importuję potrzebne biblioteki do przetwarzania obrazów wraz z rozwiązaniem problemu związanym ze złym działaniem programu imshow

```python
#Autor: Hubert Dziurla
import cv2;
from google.colab.patches import cv2_imshow
import numpy as np
```

Wczytuje obraz pełny

```python
imgload = cv2.imread('TestX1.jpg');
```

Zamieniam go na obraz czarno-biały i binaryzuje go z użyciem odpowiedniego progu binaryzacji

```python
obraz1 =  cv2.cvtColor(imgload,cv2.COLOR_RGB2GRAY);
(thresh,obraz1) = cv2.threshold(obraz1,64,255,0)
```

Tworzę element strukturalny do wykonywania operacji morfologicznych na obrazie binarnym

```python
element_strukturalny = np.ones((5),np.uint8)
```

Wykonuję operację dylatacji

```python
dylat1 = cv2.dilate(obraz1,element_strukturalny,iterations =  73)
```

Wyszukaj konturu w obrazie dylatacji i rysuję go w obrazie głównym

```python
contours = cv2.findContours(dylat1, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)[0]
imgload = cv2.drawContours(imgload,contours,-1, (0, 0, 255), 4)
cv2_imshow(imgload)
```

Zliczanie ilości wykrytych obiektów

```python
IloscObiektowWObrazie= len(contours)

print ("Wykryto: ", str(IloscObiektowWObrazie)," drzwi.")
```
